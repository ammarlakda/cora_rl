```python
def calculate_stat(rewards):
    mean_reward = np.mean(rewards)
    variance_reward = np.var(rewards, ddof=1)  # Using Bessel's correction (ddof=1)
    return mean_reward, variance_reward

pulp_mean, pulp_variance1 = calculate_stat(cumulative_alpha_4)

greedy_mean, greedy_variance1 = calculate_stat(cumulative_pulp)

# Calculate the difference in means
mean_difference = pulp_mean - greedy_mean

# Estimate the variance of the difference
# If the same sample paths are used for both policies, we need to calculate covariance
covariance1 = np.cov(cumulative_pulp, cumulative_alpha_4)[0][1]
variance_difference1 = pulp_variance1 + greedy_variance1 - 2 * covariance1

print()
print()
# print("USING NUMPY's VARIANCE FUNCTION:")

print("Alpha 4 mean:",pulp_mean)
print("Alpha 0 mean:",greedy_mean)
print("Mean Difference:",mean_difference)
# print("Variance difference:", variance_difference1)

###########################################################################################



# # Calculate the variance of the sample mean using the provided formula
def calculate_variance(rewards, mean):
    sum_of_squares = np.sum((rewards - mean) ** 2)
    variance = (1 / EPS) * (1 / (EPS - 1)) * sum_of_squares
    return variance

pulp_variance2 = calculate_variance(cumulative_pulp, pulp_mean)
greedy_variance2 = calculate_variance(cumulative_alpha_4, greedy_mean)

# variance_difference2 = pulp_variance2 + greedy_variance2 - (2 * covariance1/N)

print()
print()
print("Variance for Alpha 0:",pulp_variance2)
print("Variance for Alpha 4:", greedy_variance2)

delta_hat = list(np.array(cumulative_pulp) - np.array(cumulative_alpha_4))
prob = 1/EPS

delta_bar_list = [i * prob for i in delta_hat]
delta_bar = sum(delta_bar_list)
#std_dev_diff = np.sqrt(pulp_variance2)
# print(delta_bar)
def calculate_variance(hat, bar):
    sum_of_squares = np.sum((hat - bar) ** 2)
    variance = (1 / EPS) * (1 / (EPS - 1)) * sum_of_squares
    return variance
var = calculate_variance(delta_hat,delta_bar)
print("Variance on Mean Difference:", var)
# Assuming normality
# Standard normal deviate for a 95% confidence interval is 1.96
z_score = stats.norm.ppf(0.05)  # one-tailed test

# Margin of error
margin_of_error = z_score * np.sqrt(var) #np.sqrt(2*pulp_variance2)

# Confidence interval
confidence_interval = (mean_difference - margin_of_error, mean_difference + margin_of_error)
print()
print()
print("Confidence Interval:",confidence_interval)
```
